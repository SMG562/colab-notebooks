# loading a Transformer model that is already trained

```python
from transformers import BertModel
# instantiate a model with pretrained weights, or checkpoints
model = BertModel.from_pretrained("bert-base-cased")
```

The **identifier** used to load the model can be the identifier of any model on the Model Hub, as long as it is compatible with the BERT architecture.

```python
sequences = ["Hello!", "Cool.", "Nice!"]
```

```python
from transformers import AutoTokenizer

checkpoint = "bert-base-cased"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
encoded_sequences = tokenizer(sequences,return_tensors="pt")
```

```python
outputs = model(**encoded_sequences)
```
