# apply regular expression to a column

```python
import re

def sub_str(doi):
    return re.match('.+?(?=/)', doi).group(0)
```

```python

# method 1
df_short['doi_short'] = df_short['doi'].apply(sub_str)
```

```python
# method 2
df_short['doi_short2'] = df_short['doi'].apply(lambda x: sub_str(x))
```
