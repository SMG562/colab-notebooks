```python
re.match(pattern, str1).group(0)

# ref 
# How do I return a string from a regex match in python? - Stack Overflow
# https://stackoverflow.com/questions/18493677/how-do-i-return-a-string-from-a-regex-match-in-python
```

```python
import re

def sub_str(doi):
    return re.match('.+?(?=/)', doi).group(0)

df_short['doi_short'] = df_short['doi'].apply(sub_str)
df_short
```

```python
doi_features = pd.get_dummies(df_short['doi_short'])
```

```python
df_dois = df_short.join(doi_features)
```

```python
df_dois
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df_dois['theme_id'] = le.fit_transform(df['theme'])
```

```python
df_dois
```

```python
X = df_dois.iloc[: ,3:-1]
X
```

```python
y = df_dois.iloc[: ,-1]
y
```

# split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=27)
```

```python
X_train
```

```python
y_train
```

# classifier -knn

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN_model = KNeighborsClassifier(n_neighbors=5)
```

```python
KNN_model.fit(X_train, y_train)
```

```python
# Train set score
```

```python
KNN_model.score(X_train, y_train)
```

```python
# Test set score
```

```python
y_pred_KNN=KNN_model.predict(X_test)
```

```python
KNN_model.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

print(accuracy_score(y_pred_KNN, y_test))
```
