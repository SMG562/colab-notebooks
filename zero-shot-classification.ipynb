```python
import transformers

from transformers import pipeline

classifier = pipeline('zero-shot-classification')
```

```python
classifier(
    'Zero-Shot Learning in Modern NLP, State-of-the-art NLP models for text classification without annotated data',
    candidate_labels = ['education', 'technology', 'politics']
)
```

```python
classifier(
    'Zero-Shot Learning in Modern NLP, State-of-the-art NLP models for text classification without annotated data',
    candidate_labels = ['education', 'technology', 'politics']
)
```

```python
classifier(
    'Nick Kyrgios devastated after US Open quarterfinal defeat to Karen Khachanov',
    candidate_labels = ['education','sports', 'tennis' ,'technology', 'politics']
)
```

```python
classifier_pipeline = pipeline ("zero-shot-classification", model = "facebook/bart-large-mnli")
```

```python
classifier_pipeline(
    'Nick Kyrgios devastated after US Open quarterfinal defeat to Karen Khachanov',
    candidate_labels = ['education','sports', 'tennis' ,'technology', 'politics'],multi_label=True
)
```
