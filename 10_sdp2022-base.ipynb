# which encoders to use

- 当分类变量存在顺序关系（有序变量）时，如【小学，初中，高中】，可以使用 `sklearn.preprocessing.OrdinalEncoder` 进行编码

- 当分类变量不存在顺序关系时，如【红，黄，蓝，绿】，可以使用 `sklearn.preprocessing.OneHotEncoder` 进行编码

# 为一个列表构建词典

```python
src = ["red", "blue", "green", "red", "red", "blue", "yellow"]
color_dict = dict()
for i in range(len(src)):
  k = src[i]  # like "red"
  if k not in color_dict:
    color_dict[k] = len(color_dict)
print(color_dict)
```
