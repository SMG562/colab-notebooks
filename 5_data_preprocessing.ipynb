# data pre-process

```python
df = df[['title','theme']]
df
```

```python
df.info()
```

# label name to id

```python
topic_list = df.theme.unique()
print("Topics:\n", topic_list)
```

```python
df["theme"] = df["theme"].astype('category')
df.dtypes
```

```python
df["theme_id"] = df["theme"].cat.codes
df.tail()
```

```python
df
```

# id to category

```python
id_to_category = pd.Series(df.theme.values,index=df.theme_id).to_dict()
id_to_category
```

# category_to_id

```python
category_to_id= {v:k for k,v in id_to_category.items()}
category_to_id
```

# train_test_split

```python
# prepare feature and target
features, targets = df['title'], df['theme_id']
```

```python
# scikit-learn
from sklearn.model_selection import train_test_split
train_features, val_features, train_targets, val_targets  = train_test_split(
        features, targets,
        train_size=0.75,
        test_size=0.25,
        random_state=42,
        shuffle = True,
        stratify=targets
    )
```
